```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_1960_onwards = pd.read_csv('/content/drive/MyDrive/quickview of price surge in nigeria/1960_onwards.csv')

# Display the first 5 rows
print(df_1960_onwards.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df_1960_onwards.info())

# Convert the `Year` column to datetime format
df_1960_onwards['Year'] = pd.to_datetime(df_1960_onwards['Year'], format='%Y')

# Filter the dataframe to include only the specified columns
filtered_df = df_1960_onwards[['GDP (current LCU)', 'Inflation, consumer prices (annual %)', 'Inflation, GDP deflator (annual %)', 'Population, total', 'Crude oil price(per barrel in $)', 'Petrol Price (Naira)', 'Year']]

# Convert columns to numeric format
for col in ['GDP (current LCU)', 'Inflation, consumer prices (annual %)', 'Inflation, GDP deflator (annual %)', 'Population, total', 'Crude oil price(per barrel in $)', 'Petrol Price (Naira)']:
    filtered_df[col] = pd.to_numeric(filtered_df[col], errors='coerce')

# Fill in missing values with column means
filtered_df.fillna(filtered_df.mean(), inplace=True)
```

```python
import altair as alt

# Create a line plot of `GDP (current LCU)` by `Year`
chart1 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('GDP (current LCU):Q', axis=alt.Axis(title='GDP (current LCU)')),
    tooltip = ['Year', 'GDP (current LCU)']
).properties(
    title='GDP (current LCU) by Year'
).interactive()

# Create a line plot of `Inflation, consumer prices (annual %)` by `Year`
chart2 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('Inflation, consumer prices (annual %):Q', axis=alt.Axis(title='Inflation, consumer prices (annual %)')),
    tooltip = ['Year', 'Inflation, consumer prices (annual %)']
).properties(
    title='Inflation, consumer prices (annual %) by Year'
).interactive()

# Create a line plot of `Inflation, GDP deflator (annual %)` by `Year`
chart3 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('Inflation, GDP deflator (annual %):Q', axis=alt.Axis(title='Inflation, GDP deflator (annual %)')),
    tooltip = ['Year', 'Inflation, GDP deflator (annual %)']
).properties(
    title='Inflation, GDP deflator (annual %) by Year'
).interactive()

# Create a line plot of `Population, total` by `Year`
chart4 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('Population, total:Q', axis=alt.Axis(title='Population, total')),
    tooltip = ['Year', 'Population, total']
).properties(
    title='Population, total by Year'
).interactive()

# Create a line plot of `Crude oil price(per barrel in $)` by `Year`
chart5 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('Crude oil price(per barrel in $):Q', axis=alt.Axis(title='Crude oil price(per barrel in $)')),
    tooltip = ['Year', 'Crude oil price(per barrel in $)']
).properties(
    title='Crude oil price(per barrel in $) by Year'
).interactive()

# Create a line plot of `Petrol Price (Naira)` by `Year`
chart6 = alt.Chart(filtered_df).mark_line().encode(
    x=alt.X('Year:T', axis=alt.Axis(title='Year')),
    y=alt.Y('Petrol Price (Naira):Q', axis=alt.Axis(title='Petrol Price (Naira)')),
    tooltip = ['Year', 'Petrol Price (Naira)']
).properties(
    title='Petrol Price (Naira) by Year'
).interactive()

# Display all plots
chart1.save('gdp_current_lcu_by_year_line_plot.json')
chart2.save('inflation_consumer_prices_annual_by_year_line_plot.json')
chart3.save('inflation_gdp_deflator_annual_by_year_line_plot.json')
chart4.save('population_total_by_year_line_plot.json')
chart5.save('crude_oil_price_per_barrel_in_$_by_year_line_plot.json')
chart6.save('petrol_price_naira_by_year_line_plot.json')
```

```python
import pandas as pd
import plotly.express as px

# Read the CSV file into a DataFrame
df_1960_onwards = pd.read_csv('/content/drive/MyDrive/quickview of price surge in nigeria/1960_onwards.csv')

# Display the first 5 rows
print(df_1960_onwards.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types
print(df_1960_onwards.info())

# Convert the `Year` column to datetime format
df_1960_onwards['Year'] = pd.to_datetime(df_1960_onwards['Year'], format='%Y')

# Filter the dataframe to include only the specified columns
filtered_df = df_1960_onwards[['GDP (current LCU)', 'Inflation, consumer prices (annual %)', 'Inflation, GDP deflator (annual %)', 'Population, total', 'Crude oil price(per barrel in $)', 'Petrol Price (Naira)', 'Year']]

# Convert the columns to numeric format, setting failed conversions to missing
for col in ['GDP (current LCU)', 'Inflation, consumer prices (annual %)', 'Inflation, GDP deflator (annual %)', 'Population, total', 'Crude oil price(per barrel in $)', 'Petrol Price (Naira)']:
    filtered_df[col] = pd.to_numeric(filtered_df[col], errors='coerce')

# Fill in missing values with column means
filtered_df.fillna(filtered_df.mean(), inplace=True)

# Create and display a line plot of `GDP (current LCU)` by `Year`
fig1 = px.line(filtered_df, x='Year', y='GDP (current LCU)', title='GDP (current LCU) by Year')
fig1.show()

# Create and display a line plot of `Inflation, consumer prices (annual %)` by `Year`
fig2 = px.line(filtered_df, x='Year', y='Inflation, consumer prices (annual %)', title='Inflation, consumer prices (annual %) by Year')
fig2.show()

# Create and display a line plot of `Inflation, GDP deflator (annual %)` by `Year`
fig3 = px.line(filtered_df, x='Year', y='Inflation, GDP deflator (annual %)', title='Inflation, GDP deflator (annual %) by Year')
fig3.show()

# Create and display a line plot of `Population, total` by `Year`
fig4 = px.line(filtered_df, x='Year', y='Population, total', title='Population, total by Year')
fig4.show()

# Create and display a line plot of `Crude oil price(per barrel in $)` by `Year`
fig5 = px.line(filtered_df, x='Year', y='Crude oil price(per barrel in $)', title='Crude oil price(per barrel in $) by Year')
fig5.show()

# Create and display a line plot of `Petrol Price (Naira)` by `Year`
fig6 = px.line(filtered_df, x='Year', y='Petrol Price (Naira)', title='Petrol Price (Naira) by Year')
fig6.show()

# Create and display a scatter plot of `Crude oil price(per barrel in $)` by `Petrol Price (Naira)`
fig7 = px.scatter(filtered_df, x='Crude oil price(per barrel in $)', y='Petrol Price (Naira)', title='Crude oil price(per barrel in $) by Petrol Price (Naira)')
fig7.show()

# Create and display a scatter plot of `Inflation, consumer prices (annual %)` by `Inflation, GDP deflator (annual %)`
fig8 = px.scatter(filtered_df, x='Inflation, consumer prices (annual %)', y='Inflation, GDP deflator (annual %)', title='Inflation, consumer prices (annual %) by Inflation, GDP deflator (annual %)')
fig8.show()
```

```python
# Convert the `Year` column to datetime format
df_1960_onwards['Year'] = pd.to_datetime(df_1960_onwards['Year'], format='%Y')

# Filter the dataframe to include only the specified columns
filtered_df = df_1960_onwards[['GDP (constant LCU)', 'GDP (current LCU)', 'GDP growth (annual %)', 'GDP per capita (current US$)', 'Inflation, GDP deflator (annual %)', 'Population, total', 'Year']]

# Exclude the year 2024 from the analysis
filtered_df = filtered_df[filtered_df['Year'] != '2024-01-01']

# Convert the columns to numeric format, setting failed conversions to missing
for col in ['GDP (constant LCU)', 'GDP (current LCU)', 'GDP growth (annual %)', 'GDP per capita (current US$)', 'Inflation, GDP deflator (annual %)', 'Population, total']:
    filtered_df[col] = pd.to_numeric(filtered_df[col], errors='coerce')

# Fill in missing values with column means
filtered_df.fillna(filtered_df.mean(), inplace=True)

# Reshape the dataframe from wide to long format
filtered_df = filtered_df.melt(id_vars='Year', var_name='Indicator', value_name='Value')

# Create and display a line plot using plotly express
fig = px.line(filtered_df, x='Year', y='Value', color='Indicator', title='Economic Indicators Over Time')

# Add title and labels to the plot
fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Value',
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["color", filtered_df['Indicator'].unique()],
                    label="All",
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][0]]],
                    label=filtered_df['Indicator'][0],
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][1]]],
                    label=filtered_df['Indicator'][1],
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][2]]],
                    label=filtered_df['Indicator'][2],
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][3]]],
                    label=filtered_df['Indicator'][3],
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][4]]],
                    label=filtered_df['Indicator'][4],
                    method="update"
                ),
                dict(
                    args=["color", [filtered_df['Indicator'][5]]],
                    label=filtered_df['Indicator'][5],
                    method="update"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)

# Show the plot
fig.show()
```
