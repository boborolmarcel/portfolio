```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '1980_onwards.csv'
data = pd.read_csv(file_path)

data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Crop production'], marker='o', linestyle='-', color='b', label='Crop production')
plt.title('Crop Production Over Years')
plt.xlabel('Year')
plt.ylabel('Crop Production')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# data_columns contains the list of column names to visualize excluding the first column
data_columns = [col for col in data.columns[1:]]

plt.figure(figsize=(10, 6))

# Plotting for each column
for column in data_columns:
    plt.plot(data['Year'], data[column], marker='o', linestyle='-', color='b', label=f'{column}')
    plt.xlabel('Year')
    plt.ylabel(f'{column} Over Years')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Year'], data['Construction Sector Naira(million)'], color='g', alpha=0.7, label='Construction Sector Naira(million)')
plt.title('Construction Sector Over Years')
plt.xlabel('Year')
plt.ylabel('Construction Sector (millions in naira)')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plotting Livestock
plt.subplot(1, 2, 1)
plt.bar(data['Year'], data['Livestock'], color='b', alpha=0.7, label='Livestock')
plt.title('Livestock Over Years')
plt.xlabel('Year')
plt.ylabel('Livestock')
plt.legend()
plt.grid(axis='y')
```

```python
plt.figure(figsize=(12, 6))

# Plotting Forestry
plt.subplot(1, 2, 2)
plt.bar(data['Year'], data['Forestry'], color='g', alpha=0.7, label='Forestry')
plt.title('Forestry Over Years')
plt.xlabel('Year')
plt.ylabel('Forestry')
plt.legend()
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# data_columns contains the list of column names to visualize excluding the first column
data_columns = [col for col in data.columns[1:]]

plt.figure(figsize=(50, 50))  # Increase figure size

# Increase font size globally
plt.rcParams.update({'font.size': 18})

# Loop through each column name in data_columns and create subplots
for idx, column in enumerate(data_columns, start=1):
    plt.subplot(5, 3, idx)  # Adjust the layout based on the number of columns you have
    plt.bar(data['Year'], data[column], alpha=0.7, label=column)
    plt.title(f'{column} Over Years', fontsize=30)
    plt.xlabel('Year', fontsize=30)
    plt.ylabel(column, fontsize=26)
    plt.legend()
    plt.grid(axis='y')

    # Optionally rotate x-axis labels
    plt.xticks(rotation=45)

plt.tight_layout()
plt.suptitle("Plotting all the Columns",y=1.02, fontsize=40)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Exports of goods, fob BoP(cur. US $)'], bins=5, edgecolor='black', alpha=0.7)
plt.xlabel('Value (cur. US $)')
plt.ylabel('Frequency')
plt.title('Exports of Goods')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Lending interest rate (%)'], bins=5, edgecolor='black', alpha=0.7)
plt.xlabel('Interest Rates (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Lending Interest Rates')
plt.grid(True)
plt.show()
```

```python
data_columns = [col for col in data.columns[1:]]
# Plotting histograms for each column
for column in data_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(data[column], bins=5, edgecolor='black', alpha=0.7)
    plt.xlabel(f'{column}', fontsize=16)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}', fontsize=20)
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Account_Balance Billions of USD'], data['Exports of goods, fob BoP(cur. US $)'], color='r', marker='o', alpha=0.8)
plt.title('Account Balance vs. Exports of Goods')
plt.xlabel('Account Balance (Billions of USD)')
plt.ylabel('Exports of Goods (cur. US $)')
plt.grid(True)
plt.show()
```

```python
# Group data by decade
data['Decade'] = data['Year'] // 10 * 10
grouped = data.groupby('Decade')

# Plotting each decade separately
plt.figure(figsize=(12, 8))

for decade, group in grouped:
    plt.scatter(group['Account_Balance Billions of USD'],
                group['Exports of goods, fob BoP(cur. US $)'],
                label=f'Year {decade}s',
                alpha=0.8)

plt.title('Account Balance vs. Exports of Goods (Segmented by Decade)')
plt.xlabel('Account Balance (Billions of USD)')
plt.ylabel('Exports of Goods (cur. US $)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
labels = ['Manufacturing, value added (% of GDP)','Agriculture, forestry, and fishing, value added (% of GDP)', 'Industry (including construction), value added (% of GDP)']
sizes = [sum(data['Manufacturing, value added (% of GDP)']), sum(data['Agriculture, forestry, and fishing, value added (% of GDP)']), sum(data['Industry (including construction), value added (% of GDP)'])]
colors = ['lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0)  # explode 1st slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Composition of GDP Sectors')
plt.axis('equal')
plt.show()
```
