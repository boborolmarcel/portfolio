```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('RealGDP1981_2023_processed (1).csv')
df.head()
```

## Shows the trend and changes in GDP over time

```python
#
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Time'], y=df['Services'], marker='o', linestyle='--', color='b', label='Services')
sns.lineplot(x=df['Time'],y=df['Agriculture'], marker='o', linestyle='--', color='g', label='Agriculture')
plt.title('GDP Over Time for Services and Agriculture in Nigeria')
plt.xlabel('Year')
plt.ylabel('GDP(in million Naira) ')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

* The chart shows how Nigeria's Gross Domestic Product (GDP) has changed from 1981 to 2023 for services and agriculture.
* Key Points:
    - Services Sector:  The blue squares show that the GDP for services has greatly increased after 2000, meaning this sector has grown a lot.
    - Agriculture Sector: The green diamonds show that agriculture GDP has also increased, but not as quickly as services.

```python
# Agricultural Production Over Time
plt.figure(figsize=(10, 6))

plt.plot(df['Time'], df['CropProduction'], label='Crop production')
plt.plot(df['Time'], df['Livestock'], label='Livestock')
plt.plot(df['Time'], df['Forestry'], label='Forestry')
plt.plot(df['Time'], df['Fishing'], label='Fishing')
plt.title('Agricultural Production Over Time')
plt.xlabel('Year')
plt.ylabel('Production Volume')
plt.legend()
plt.grid(True)
plt.show()
```

* Crop production has displayed a consistent upward trend, indicating substantial and continuous growth over time. In contrast, the livestock sector has remained stagnant, performing close to the bottom of the graph compared to crop production. Furthermore, the forestry and fishing industries have also shown minimal growth when compared to crop production's upward trajectory.

```python
# Services Over Time
plt.figure(figsize=(10, 6))

plt.plot(df['Time'], df['Trade'],label='Trade')
plt.plot(df['Time'], df['AccommodationAndFoodServices'], label='AccommodationAndFoodServices')
plt.plot(df['Time'], df['TransportationAndStorage'], label='TransportationAndStorage')
plt.plot(df['Time'], df['InformationAndCommunication'], label='InformationAndCommunication')
plt.plot(df['Time'], df['ArtsEntertainmentAndRecreation'], label='ArtsEntertainmentAndRecreation')
plt.plot(df['Time'], df['FinanceAndInsurance'], label='FinanceAndInsurance')
plt.plot(df['Time'], df['RealEstate'], label='RealEstate')
plt.plot(df['Time'], df['ProfessionalScientificAndTechnicalServices'], label='ProfessionalScientificAndTechnicalServices')
plt.plot(df['Time'], df['AdministrativeAndSupportServicesBusinessServices'], label='AdministrativeAndSupportServicesBusinessServices')
plt.plot(df['Time'], df['PublicAdministration'], label='PublicAdministration', linestyle='--')
plt.plot(df['Time'], df['Education'], label='Education', linestyle='--')
plt.plot(df['Time'], df['ProfessionalScientificAndTechnicalServices'], label='ProfessionalScientificAndTechnicalServices', linestyle='--')
plt.plot(df['Time'], df['HumanHealthAndSocialServices'], label='HumanHealthAndSocialServices', linestyle='--')
plt.plot(df['Time'], df[ 'OtherServices'], label= 'OtherServices', linestyle='--')
plt.title('Servies Over Time')
plt.xlabel('Year')
plt.ylabel('Production Volume')
plt.legend()
plt.grid(True)
plt.show()
```

* In the time period from 1981 to around 2023, the production volume of various services showed the following trends:

  - Accommodation and Food Services: demonstrated moderate growth over the years.
  - Transportation and Storage: also showed steady growth.
  - Information and Communication: experienced significant growth, especially after 2000.
  - Arts, Entertainment, and Recreation: displayed fluctuations but no substantial trend.
  - Finance and Insurance: showed moderate growth.
  - Real Estate Activities: remained relatively stable.
  - Professional, Scientific, and Technical Services: experienced steady growth.
  - Education: no clear trend was observed.
  - Public Administration: remained relatively stable.
  - Human Health and Social Services: exhibited steady growth.
  - Other Services: no significant trend was identified.

## Compares GDP values across different years

```python

plt.figure(figsize=(10, 6))
plt.bar(df['Time'], df['Services'], color='skyblue')
plt.title('Services in Nigeria')
plt.xlabel('Year')
plt.ylabel('GDP (in million Naira)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(df['Time'], df['Agriculture'], color='skyblue')
plt.title('Agricultural Production in Nigeria')
plt.xlabel('Year')
plt.ylabel('GDP (in million Naira)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Illustrates the composition or cumulative aspects of GDP components over time.

```python
plt.figure(figsize=(10, 6))
plt.stackplot(df['Time'],df['Agriculture'], df['Services'], labels=['GDP at 2010 Constant Market Prices','Agriculture', 'Services'])
plt.title('Sector-wise GDP Composition Over Time')
plt.xlabel('Year')
plt.ylabel('GDP (in million Naira)')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

* Services Sector: Services also contribute to GDP growth, although at a slower rate.
* Agriculture Sector: Agriculture has the least growth in GDP contribution.
* Period: The data spans from 1981 to approximately 2023.

```python
import numpy as np
years=df['Time']
# Example data (replace with actual data loading)
years = np.arange(len(years))  # Generating indices for years
bar_width = 0.35  # Width of the bars

plt.figure(figsize=(10, 6))
#plt.bar(years - bar_width/2, df['GDP at 2010 Constant Market Prices'], bar_width, label='GDP at 2010 Constant Market Prices')
plt.bar(years - bar_width/2, df['Services'], bar_width, label='Services')
plt.bar(years + bar_width/2, df['Agriculture'], bar_width, label='Agriculture')
plt.title('GDP Comparison for Services and Agriculture in Nigeria')
plt.xlabel('Year')
plt.ylabel('GDP (in million Naira)')
plt.xticks(years, [str(year) for year in years])
plt.legend()
plt.tight_layout()
plt.show()
```

The GDP Comparison for Services and Agriculture in Nigeria provides data on the GDP (in millions of Naira) for the services and agriculture sectors over a span of 41 years. 
Key points to note: 
 - Services GDP: The blue bars show the GDP for the services sector, which has demonstrated steady growth over time. 
 - Agriculture GDP: The orange bars represent the GDP for agriculture, which has consistently remained lower compared to services. 
 - Trend: The data illustrates that services have become a larger component of Nigeria's economy compared to agriculture.
