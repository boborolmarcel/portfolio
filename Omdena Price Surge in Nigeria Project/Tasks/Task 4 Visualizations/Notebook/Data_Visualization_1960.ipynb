```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)



import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Verify the file path and try listing files in the directory
!ls "/content/drive/My Drive/Data Visualization/"

# Read the data from the CSV file
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/1960_onwards.csv');


# Convert the 'Year' column to datetime type
df['Year'] = pd.to_datetime(df['Year'], format='%Y')

### GDP growth rate vs. inflation rate

plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP growth (annual %)', y='Inflation, consumer prices (annual %)', data=df)
plt.xlabel('GDP growth rate')
plt.ylabel('Inflation rate')
plt.title('GDP growth rate vs. inflation rate')

### Population distribution

plt.figure(figsize=(10, 6))
sns.barplot(x='Population ages 0-14 (% of total population)', y='Population ages 15-64 (% of total population)', data=df)
plt.xlabel('Population ages 0-14 (%)')
plt.ylabel('Population ages 15-64 (%)')
plt.title('Population distribution')

### Money supply

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Narrow Money', data=df, label='Narrow Money')
sns.lineplot(x='Year', y='Money Supply M3', data=df, label='Money Supply M3')
sns.lineplot(x='Year', y='Net Domestic Assets', data=df, label='Net Domestic Assets')
sns.lineplot(x='Year', y='Money Supply M2', data=df, label='Money Supply M2')
plt.xlabel('Year')
plt.ylabel('Amount (billions of US $)')
plt.title('Money supply')
plt.legend()

### Crude oil production

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Cumulative crude oil production up to and including year', data=df)
plt.xlabel('Year')
plt.ylabel('Cumulative crude oil production (millions of barrels)')
plt.title('Crude oil production')

### Petrol price

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Petrol Price (Naira)', data=df)
plt.xlabel('Year')
plt.ylabel('Petrol price (Naira)')
plt.title('Petrol price')

### Trade openness

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Trade Openness Index(%)', data=df)
plt.xlabel('Year')
plt.ylabel('Trade openness index (%)')
plt.title('Trade openness')

plt.show()
### Demand Deposits

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Demand Deposits', data=df)
plt.xlabel('Year')
plt.ylabel('Demand Deposits')
plt.title('Demand Deposits')

### Quasi Money

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Quasi Money', data=df)
plt.xlabel('Year')
plt.ylabel('Quasi Money')
plt.title('Quasi Money')

### Net Domestic Assets

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Net Domestic Assets', data=df)
plt.xlabel('Year')
plt.ylabel('Net Domestic Assets')
plt.title('Net Domestic Assets')

### Other Assets Net

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Other Assets Net', data=df)
plt.xlabel('Year')
plt.ylabel('Other Assets Net')
plt.title('Other Assets Net')

### Money Supply M2

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Money Supply M2', data=df)
plt.xlabel('Year')
plt.ylabel('Money Supply M2')
plt.title('Money Supply M2')

### CBN Bills

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='CBN Bills', data=df)
plt.xlabel('Year')
plt.ylabel('CBN Bills')
plt.title('CBN Bills')

### Special Intervention Reserves

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Special Intervention Reserves', data=df)
plt.xlabel('Year')
plt.ylabel('Special Intervention Reserves')
plt.title('Special Intervention Reserves')

### GDP Billions of US $

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='GDPBillions of US $', data=df)
plt.xlabel('Year')
plt.ylabel('GDP Billions of US $')
plt.title('GDP Billions of US $')

### Per Capita US $

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Per CapitaUS $', data=df)
plt.xlabel('Year')
plt.ylabel('Per Capita US $')
plt.title('Per Capita US $')

### Growth Rate Annual % Change

plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Growth RateAnnual % Change', data=df)
plt.xlabel('Year')
plt.ylabel('Growth Rate Annual % Change')
plt.title('Growth Rate Annual % Change')

plt.show()
```
