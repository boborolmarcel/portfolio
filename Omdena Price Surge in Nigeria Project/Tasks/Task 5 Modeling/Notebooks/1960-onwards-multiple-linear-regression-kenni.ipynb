Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
data = pd.read_csv('1960_onwards.csv')
```

```python
# number of rows and columns in the dataset
data.shape
```

```python
# statistical measures about the data
data.describe()
```

```python
# checking the distribution of Target Variable
data['GDP per capita (current US$)'].value_counts()
```

Splitting the Features and Target

```python
X = data.drop(columns='GDP per capita (current US$)', axis=1) #if row axis=0
Y = data['GDP per capita (current US$)']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data (80% 20%)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Linear Regression

```python
model = LinearRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
MAE=mean_absolute_error(Y_test,y_pred)
```

```python
MAE
```

```python
mean_absolute_percentage_error(Y_test,y_pred)
```

```python
MSE=mean_squared_error(Y_test,y_pred)
```

```python
MSE
```

```python
r2 = r2_score(Y_test, y_pred)
print(f'R^2 score: {r2}')
```

```python
RMSE= np.sqrt(MSE)
```

```python
RMSE
```

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(Y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('y_test', fontsize=18)                          # X-label
plt.ylabel('y_pred', fontsize=16)    
```
