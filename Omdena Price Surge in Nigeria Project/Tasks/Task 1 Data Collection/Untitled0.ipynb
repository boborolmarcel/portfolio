```python
import os

def count_files_in_folder(folder_path):
    """Counts the number of files in a specified folder."""

    count = 0
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            count += 1

    return count

# Example usage
folder_to_count = "your_folder_path_here"  # Replace with the actual folder path
file_count = count_files_in_folder(folder_to_count)
print(f"There are {file_count} files in the folder '{folder_to_count}'.")
```
