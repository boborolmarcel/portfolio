# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv = pd.read_csv("bestsellers.csv")
csv.info()
```

Are the number of reviews increasing each year?
Which book received the most numbers of reviews?

```python
csv.isnull().sum()
```

```python
csv[csv["Name"].isna()]
```

```python
csv = csv.dropna()
csv.isnull().sum()
```

```python
 csv.info()
```

```python
csv.head()
```

```python
csv["Year"] = csv["Year"].astype(int)
csv.info()
```

```python
csv.head()
```

```python
csv = csv.sort_values(by = "Reviews", ascending = False)
csv = csv.head(10)
csv
```

```python
s_trend = csv.groupby("Year", as_index = False).agg({"Reviews": "sum"})
```

```python
del csv["Type"]
csv.head()
```

plotting the result

Are the nuber of reviews increasing each year?

```python
sns.lineplot(x = "Year", y = "Reviews", data = s_trend, marker="o" )
plt.title("Trend of Reviews")
plt.ylabel("Sales")
plt.show()
```

 Which book received the most number of reviews?

```python
sns.barplot(x = "Reviews", y= "Name", data= csv, ci= False, orientation="horizontal", palette="Blues")
plt.xlabel("")
plt.ylabel("")
plt.title("Review of Books")
plt.show()
```
